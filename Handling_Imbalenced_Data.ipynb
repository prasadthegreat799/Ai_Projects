Imbalenced Dataset:A Dataset with unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('/content/credit_data.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.isnull()
```

```python
#distribution of two classes
dataset['Class'].value_counts()
```

This is imbalanced Dataset

0-->legit Transactions
1-->Fraud Transactions

```python
#seperating the legit and fraud transactions
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

UnderSampling

Building a sample dataset containging the similar dirstubution of legit and fraud transactions.

```python
legit_sample=legit.sample(n=201)
```

```python
legit_sample.shape
```

Concatinating the fraud and legit datasets

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
#here axis=0 row wise
```
